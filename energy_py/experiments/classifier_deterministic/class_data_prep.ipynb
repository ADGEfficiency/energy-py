```python
import pandas as pd
```

```python
actual_price = pd.read_csv('actual_price.csv', index_col=0, parse_dates=True)
actual_price.head()
```

```python
actual_price.columns = ['C_electricity_price_[$/MWh]']
```

```python
p = pd.read_csv('price_predictions.csv', index_col=0, parse_dates=True)
p = p.drop('Current_price', axis=1)
```

```python
p.head()
```

```python
unencoded = [p.iloc[:, n] for n in range(p.shape[1])]
```

```python
encoded = [pd.get_dummies(col) for col in unencoded]
names = ['h_6_', 'h_12_', 'h_24_']
encoded = [df.add_prefix(name) for df, name in zip(encoded, names)]
```

```python
obs = pd.concat(encoded, axis=1)
obs.head()
```

```python
p_current = pd.read_csv('HH_horizon_prediction_encoded.csv', index_col=0, parse_dates=True)
p_current.head()
```

```python
p_current.columns = ['h_0_High', 'h_0_Low', 'h_0_Medium', 'h_0_Negative', 'h_0_Very High']
p_current.head()
```

```python
obs.head()
```

```python
observation = pd.concat([p_current, obs], axis=1).dropna()
observation.head()
```

```python
state, observation = actual_price.align(observation, axis=0, join='inner')
```

```python
state.head()
```

```python
observation.head()
```

```python
assert state.index[0] == observation.index[0]
assert state.index[-1] == observation.index[-1]
```

```python
state.to_csv('state.csv')
observation.to_csv('observation.csv')
```
